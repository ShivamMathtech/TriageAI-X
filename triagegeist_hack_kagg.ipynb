# Clinical Text Cleaning

Clinical text preprocessing includes:
- lowercase conversion
- punctuation removal
- whitespace normalization

```python
# ==========================================
# TEXT CLEANING FUNCTION
# ==========================================

import re

def clean_text(text):
    
    text = str(text).lower()
    
    text = re.sub(
        r"[^a-zA-Z0-9\s]",
        "",
        text
    )
    
    text = re.sub(
        r"\s+",
        " ",
        text
    ).strip()
    
    return text
```

```python
# ==========================================
# CLEAN COMPLAINT TEXT
# ==========================================

train_df["clean_complaint"] = (
    train_df["chief_complaint_raw"]
    .apply(clean_text)
)

train_df[
    [
        "chief_complaint_raw",
        "clean_complaint"
    ]
].head()
```

```python
# ==========================================
# COMPLAINT LENGTH
# ==========================================

train_df["complaint_length"] = (
    train_df["clean_complaint"]
    .apply(len)
)

train_df[
    "complaint_length"
].describe()
```

# TF-IDF Vectorization

TF-IDF converts clinical text into numerical representations.

This enables machine learning models to learn:
- symptom patterns
- emergency terminology
- severity-related language

```python
# ==========================================
# TF-IDF VECTORIZATION
# ==========================================

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=300,
    stop_words="english"
)

tfidf_matrix = tfidf.fit_transform(
    train_df["clean_complaint"]
)

print("TF-IDF Completed")
```

```python
# ==========================================
# TF-IDF MATRIX SHAPE
# ==========================================

tfidf_matrix.shape
```

```python
# ==========================================
# TF-IDF DATAFRAME
# ==========================================

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df.head()
```

```python
# ==========================================
# MERGE NLP FEATURES
# ==========================================

train_df = pd.concat(
    [
        train_df.reset_index(drop=True),
        tfidf_df.reset_index(drop=True)
    ],
    axis=1
)

print(train_df.shape)
```
