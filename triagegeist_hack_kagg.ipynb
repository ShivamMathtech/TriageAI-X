# Phase 5 — Data Integration

This phase combines:
- Structured patient data
- Chief complaints

using:

```python
# ==========================================
# MERGE DATASETS
# ==========================================

train_df = train_df.merge(
    complaints_df,
    on="patient_id",
    how="left"
)

train_df = train_df.merge(
    history_df,
    on="patient_id",
    how="left"
)

print("Merged Dataset Shape :", train_df.shape)
```

# Phase 6 — Data Preprocessing

This phase prepares the dataset for machine learning.

Tasks include:
- Missing value handling
- Encoding
- Normalization
- Text cleaning

```python
# ==========================================
# HANDLE MISSING VALUES
# ==========================================

for col in train_df.columns:
    
    if train_df[col].dtype == "object":
        train_df[col] = train_df[col].fillna("Unknown")
        
    else:
        train_df[col] = train_df[col].fillna(
            train_df[col].median()
        )

print("Missing Values Handled")
```

# Encoding Categorical Variables

```python
# ==========================================
# LABEL ENCODING
# ==========================================

encoder = LabelEncoder()

categorical_cols = train_df.select_dtypes(
    include="object"
).columns

for col in categorical_cols:
    
    train_df[col] = encoder.fit_transform(
        train_df[col]
    )

print("Categorical Variables Encoded")
```

# Feature Scaling

```python
# ==========================================
# FEATURE SCALING
# ==========================================

scaler = StandardScaler()

numerical_cols = train_df.select_dtypes(
    include=np.number
).columns

train_df[numerical_cols] = scaler.fit_transform(
    train_df[numerical_cols]
)

print("Features Scaled")
```

# Phase 7 — Clinical Feature Engineering

This phase generates medically meaningful features from raw physiological data.

Clinical indicators improve:
- Severity estimation
- Risk detection
- Model interpretability

# Shock Index

Shock Index is an important indicator of hemodynamic instability.

```python
# ==========================================
# SHOCK INDEX
# ==========================================

train_df["shock_index"] = (
    train_df["heart_rate"] /
    train_df["systolic_bp"]
)

train_df["shock_index"].head()
```

# Pulse Pressure

```python
# ==========================================
# PULSE PRESSURE
# ==========================================

train_df["pulse_pressure"] = (
    train_df["systolic_bp"] -
    train_df["diastolic_bp"]
)

train_df["pulse_pressure"].head()
```

# Hypoxia Indicator

Hypoxia is detected when oxygen saturation falls below safe thresholds.

```python
# ==========================================
# HYPOXIA FLAG
# ==========================================

train_df["hypoxia_flag"] = (
    train_df["oxygen_saturation"] < 92
).astype(int)

train_df["hypoxia_flag"].value_counts()
```

# Tachycardia Indicator

```python
# ==========================================
# TACHYCARDIA FLAG
# ==========================================

train_df["tachycardia_flag"] = (
    train_df["heart_rate"] > 100
).astype(int)

train_df["tachycardia_flag"].value_counts()
```

```python
# ==========================================
# ENGINEERED FEATURES
# ==========================================

engineered_features = [
    "shock_index",
    "pulse_pressure",
    "hypoxia_flag",
    "tachycardia_flag"
]

train_df[engineered_features].head()
```
